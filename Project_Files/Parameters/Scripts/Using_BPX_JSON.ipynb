```python
import pybamm as pb
import bpx
from devtools import pprint
import json
```

```python
# Load the BPX JSON file as a XPS format
filename = r'C:\Users\frw21\OneDrive - Imperial College London\Misc\Desktop\ESE_UROP_2025\Project_Files\Parameters\example_BPX.json'
my_params = bpx.parse_bpx_file(filename)
```

```python
# Printing function objects in BPX as a python function
positive_electrode_diffusivity = my_params.parameterisation.electrolyte.conductivity.to_python_function()
diff_at_one = positive_electrode_diffusivity(1.0)
print('positive electrode diffusivity at x = 1.0:', diff_at_one)
```

```python
# Convert the BPX parameters to a dictionary
# This will allow you to access and modify the parameters easily
my_params_dict = my_params.model_dump(by_alias=True)
print('Initial temperature of cell:', my_params_dict["Parameterisation"]["Cell"]["Initial temperature [K]"])
my_params_dict["Parameterisation"]["Cell"]["Initial temperature [K]"]= 100
print('Initial temperature of cell:', my_params_dict["Parameterisation"]["Cell"]["Initial temperature [K]"])
```

```python
# Save the modified parameters back to a JSON file
import json

with open("example_BPX copy.json", "w") as f:
    json.dump(my_params_dict, f, indent=2)
```

```python
# Load the BPX JSON file as a XPS format
filename = r'C:\Users\frw21\OneDrive - Imperial College London\Misc\Desktop\ESE_UROP_2025\Project_Files\Parameters\example_BPX.json'
my_params = bpx.parse_bpx_file(filename)
my_params_dict = my_params.model_dump(by_alias=True)
pprint(my_params_dict)
```

```python
Param_ORegan = pb.ParameterValues("ORegan2022")
print("Old format is:")
pprint(my_params_dict)

for key, value in Param_ORegan.items():
    print(f"{key}: {value}")
    my_params_dict["Parameterisation"]["Cell"][key] = value

print("New format is:")
pprint(my_params_dict)
```

```python
import csv

Param_ORegan = pb.ParameterValues("ORegan2022")
# Save Param_ORegan to a CSV file
with open("Param_ORegan.csv", "w", newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Parameter", "Value"])
    for key, value in Param_ORegan.items():
        writer.writerow([key, value])
```
