```python
import pybamm as pb
import matplotlib.pyplot as plt
```

```python
pb.set_logging_level("NOTICE")

model = pb.equivalent_circuit.Thevenin(options={"number of rc elements": 2})
params = pb.ParameterValues("ECM_Example")

params.update(
    {
        "R2 [Ohm]": 0.0003,
        "C2 [F]": 40000,
        "Element-2 initial overpotential [V]": 0,
    },
    check_already_exists=False,
)

#Simple discharge experiment
experiment = pb.Experiment(
    [
        (
            "Discharge at 1C for 100s",
            "Rest for 250 seconds",
        )
    ]
)

# run pybamm model (LT)
sim = pb.Simulation(
    model=model, 
    experiment = experiment,
    parameter_values=params,
    solver=model.default_solver
    )
solution = sim.solve(calc_esoh=True)


solution = sim.solve(calc_esoh=True)
```

```python
plt.plot(solution["Time [s]"].data, solution["Voltage [V]"].data, label="Modelled Voltage")
plt.xlabel("Time [s]")
plt.ylabel("Voltage [V]")
plt.legend()
plt.title("Voltage Data with Noise")
plt.show()      
```

```python
# Automatically generate param_variations and original_params for all parameters in params
param_variations = {}
original_params = {}

for key, value in params.items():
    original_params[key] = value
    # Only sweep numeric parameters (int/float), skip functions or non-numeric
    if isinstance(value, (int, float)):
        # Generate a sweep: original, 0.5x, 1.5x, 2x (avoid zero)
        sweep = [value]
        if value != 0:
            sweep += [value * 0.5, value * 1.5, value * 2]
        param_variations[key] = sweep

# Remove duplicates in sweep lists
for k, v in param_variations.items():
    param_variations[k] = list(dict.fromkeys(v))

for param_name, values in param_variations.items():
    plt.figure()
    for val in values:
        # Update parameter (if original is a function, replace with a constant for sweep)
        params.update({param_name: val}, check_already_exists=False)
        sim = pb.Simulation(model=model, experiment=experiment, parameter_values=params, solver=model.default_solver)
        solution = sim.solve(calc_esoh=True)
        plt.plot(solution["Time [s]"].data, solution["Voltage [V]"].data, label=f"{param_name}={val}")
    plt.xlabel("Time [s]")
    plt.ylabel("Voltage [V]")
    plt.title(f"Effect of varying {param_name} on Voltage")
    plt.legend()
    plt.show()
    # Restore original parameter value
    params.update({param_name: original_params[param_name]}, check_already_exists=False)
```
