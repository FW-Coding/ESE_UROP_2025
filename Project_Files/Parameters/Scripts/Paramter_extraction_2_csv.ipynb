```python
import pybamm as pb
import numpy as np
import bpx
from devtools import pprint
import json
import csv
```

```python
name = "Ai2020"
param = pb.ParameterValues(name)
print(param)
```

```python
with open(name + "_parameters.csv", "w", newline='') as csvfile:
    writer = csv.writer(csvfile)
    # Write the header
    writer.writerow(["Parameter", "Value"])
    
    # Iterate through the parameters and write them to the CSV file
    for key, value in param.items():
        if isinstance(value, (int, float, np.number)):
            writer.writerow([key, value])
        elif isinstance(value, str):
            writer.writerow([key, value])
        else:
            # Convert other types to string
            writer.writerow([key, str(value)])
```
